```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd

# Print the current working directory
print("Current working directory:", os.getcwd())

```

```python
def calculate_predict_accuracy(predict, actual):
    """
    Calculate the accuracy of the prediction
    :param predict: the predicted values
    :param actual: the actual values
    :return: the accuracy of the prediction
    """
    # Calculate the accuracy of the prediction
    accuracy = np.mean(predict == actual)
    return accuracy
```

```python
predict_dir = './out_sleepedf/predict/'

# List all files in the predict directory
file_list = os.listdir(predict_dir)
print("Files in predict directory:", file_list)

# Create a pandas DataFrame to save accuracy results
accuracy_results = pd.DataFrame(columns=['File', 'Accuracy'])



```

```python

for n in file_list:
    # Load the predicted values
    predict = np.load(predict_dir + n)
    print("Predicted values:", predict)


    # Calculate the accuracy of the prediction
    accuracy = calculate_predict_accuracy(predict['y_pred'], predict['y_true'])
    new_row = pd.DataFrame({'File': [n], 'Accuracy': [accuracy]})
    accuracy_results = pd.concat([accuracy_results, new_row], ignore_index=True)
    print("Prediction accuracy:", accuracy)

    # Plot the predicted and actual values
    plt.plot(predict['y_pred'], label='Predicted')
    plt.plot(predict['y_true'], label='Actual')
    plt.title('Prediction accuracy: ' + str(accuracy))
    plt.legend()
    plt.show()

```

```python
from datetime import datetime

# Get the current date
current_date = datetime.now().strftime('%Y-%m-%d')

# Save the accuracy results to a CSV file with the current date in the file name
accuracy_results.to_csv(f'./results/accuracy_results_{current_date}.csv', index=False)

print(np.mean(accuracy_results['Accuracy']), np.std(accuracy_results['Accuracy']))


```
